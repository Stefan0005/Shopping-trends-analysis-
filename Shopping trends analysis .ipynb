DATA EXPLORATION AND VIZ

```python
#Print and with the aid of a chart viz the most popular mode of payment #results have been sorted in descending order and the first result displayed 
df.groupby('Payment Method')['Payment Method'].count().sort_values(ascending=False)
```

```python
#results have been sorted in descending order and the first result displayed 
df.groupby('Payment Method')['Payment Method'].count().sort_values(ascending=False).head(1)
```

```python
#visialization
method= df['Payment Method'].value_counts().index
base_color=sb.color_palette()[0]
sb.countplot(data=df,y='Payment Method',color=base_color,order=method)
```

```python
#Find,print and vizualise the most popular item purchased by customers 
df.groupby('Item Purchased')['Payment Method'].count().sort_values(ascending=False).head()
```

```python
#Viz
method= df['Item Purchased'].value_counts().index
base_color=sb.color_palette()[3]
sb.countplot(data=df,y='Item Purchased',color=base_color,order=method)
```

```python
#Identify and print the most purchased item by females 
df[(df['Gender']=='Female')].groupby('Item Purchased')['Item Purchased'].count().sort_values(ascending=False).head(1)
```

```python
#viz the location that drives the most sales 
method= df['Location'].value_counts().head(10).index
base_color=sb.color_palette()[]
sb.countplot(data=df,y='Location',color=base_color,order=method)
```

```python
#Average customer spend 
Average_purchases=round(df['Purchase Amount (USD)'].mean(),2)
```

```python
print('mean= '+str (Average_purchases))
```

```python
#Exploring the relationship between Age and Purchase amount using a scatterplot
#Creating a dictionary mapping genders to colors
gender_to_color = {'Male': 'blue', 'Female': 'red'}

plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Purchase Amount (USD)'], c=df['Gender'].map(gender_to_color))
plt.title('Purchase Amount vs. Age by Gender')
plt.xlabel('Age')
plt.ylabel('Purchase Amount (USD)')
plt.legend()
plt.grid(True)
plt.show()
```
